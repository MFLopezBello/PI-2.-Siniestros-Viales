## MATRIZ DE CORRELACION

```python
#selecciona las columnas numéricas del df
dataNumerica = df_homicidios.select_dtypes(include=['number']) 

#calcula la matriz de correlación
corr = dataNumerica.corr().round(2)

#configura el tamaño de la figura
plt.figure(figsize=(6,5))

# Crea un mapa de calor con seaborn
sns.heatmap(corr, cmap="summer", annot=True)

# Añade título al mapa de correlación
plt.title('Mapa de Correlación')

# Muestra el mapa de correlación
plt.show()
```

Este mapa de correlación representa las relaciones entre las variables numéricas de tu dataset, que incluye N_VICTIMAS, COMUNA, Altura y MM. La matriz de correlación es una tabla en la que cada celda muestra el coeficiente de correlación de Pearson entre dos variables. El coeficiente de correlación varía entre -1 y 1, y mide la fuerza y la dirección de la relación lineal entre dos variables.

En este mapa de correlación, puedes observar lo siguiente:

El valor en la celda de la fila N_VICTIMAS y columna COMUNA es -0.05. Esto indica una correlación débil y negativa entre el número de víctimas y la comuna. Cuanto más baja es la correlación (más cerca de 0), menos fuerte es la relación lineal entre las variables. En este caso, una correlación negativa (-0.05) sugiere que, a medida que una variable aumenta, la otra tiende a disminuir, aunque la relación es muy débil en este caso.

La celda en la fila N_VICTIMAS y columna Altura tiene un valor de -0.01. Esto también indica una correlación muy débil y negativa entre el número de víctimas y la altura de los individuos.

La celda en la fila N_VICTIMAS y columna N_VICTIMAS tiene un valor de 1. Esto es el coeficiente de correlación de una variable consigo misma y siempre será 1, lo que indica una correlación perfecta.

La celda en la fila COMUNA y columna Altura tiene un valor de -0.8. Esto indica una correlación moderadamente fuerte y negativa entre la comuna y la altura. En otras palabras, a medida que la comuna aumenta, la altura tenderá a disminuir. Sin embargo, esto no implica causalidad, ya que la correlación no prueba que una variable cause cambios en la otra.

La celda en la fila MM y columna MM tiene un valor de 1. Esto indica una correlación perfecta de la variable consigo misma.

El valor en la celda de la fila DD y columna COMUNA es 1.0. Esto significa una correlación perfecta entre las variables DD y COMUNA. Esto podría deberse a que ambas variables representan lo mismo o están estrechamente relacionadas.

En general, este mapa de correlación muestra relaciones débiles entre las variables, excepto entre COMUNA y DD, que tienen una correlación perfecta. Es importante recordar que la correlación no implica causalidad y que estos valores son solo una medida de la relación lineal entre dos variables.
